```python
Feat = ["Age","Income","Total_spending ","Recency","NumWebPurchases","NumStorePurchases","NumWebVisitsMonth"]
```

```python
x = df[Feat]
x
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler
```

```python
x_scaler = scaler.fit_transform(x)
x_scaler
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for i in range(2,10):
    kme = KMeans(n_clusters = i)
    kme.fit(x_scaler)
    wcss.append(kme.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2,10),wcss,marker = "o")
plt.title("Elbow method for Optimal K")
plt.xlabel("Number of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans  = KMeans(n_clusters = 6)
df["cluster"] = kmeans.fit_predict(x_scaler)
```

```python
df.head()
```

```python
cluster_summary = df.groupby("cluster")[Feat].mean()
cluster_summary
```

```python
palette = sns.color_palette('Paired', 6)
sns.scatterplot(data = df,x = "Age",y = "Total_spending ",hue = "cluster",palette = palette)
```

```python
df["cluster"].value_counts()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca_data = pca.fit_transform(x_scaler)
df["PCA1"],df["PCA2"] = pca_data[:,0],pca_data[:,1]
```
